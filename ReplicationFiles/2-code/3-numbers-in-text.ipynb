# Numbers in the paper's text

This file is part of the replication code for: 

Offshore Profit Shifting and Aggregate Measurement: Balance of Payments, Foreign Investment, Productivity, and the Labor Share by Guvenen, Mataloni Jr., Rassier, and Ruhl. 

This version: February 21, 2022

This file uses the files created from `2-analysis-I.ipynb`, `2-analysis-II.ipynb`, and `2-analysis-III.ipynb` named `numbers-in-text-I.csv`, `numbers-in-text-II.csv`, `numbers-in-text-III.csv`. It creates the file `numbers.csv` that is saved in `4-numbers-in-text`. 



```python
import pandas as pd

def format_numbers(n, f):
    return f.format(n)
```

```python
df1 = pd.read_csv('../3-intermediate-files/numbers-in-text-I.csv', index_col=[0])
df2 = pd.read_csv('../3-intermediate-files/numbers-in-text-II.csv', index_col=[0])
df3 = pd.read_csv('../3-intermediate-files/numbers-in-text-III.csv', index_col=[0])

numbers = pd.concat([df1, df2, df3]).sort_index()
numbers['value for paper'] = numbers[['Number', 'Format']].apply(lambda x : format_numbers(*x), axis=1)
numbers = numbers[['Description', 'value for paper']]

numbers.to_csv('../4-numbers-in-text/numbers.csv')
numbers
```
